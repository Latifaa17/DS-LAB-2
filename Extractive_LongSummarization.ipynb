```python
from summarizer import Summarizer

summarizer = Summarizer()
```

```python
import os

lectures_list = []
for root, dirs, files in os.walk("/content/data"):
    for file in files:
        if file.endswith('.txt'):
            with open(os.path.join(root, file), 'r') as f:
                article = f.read()
                lectures_list.append(article)
```

```python
from tqdm import tqdm

lectures_summaries = {}

for lecture in tqdm(lectures_list):

  summary = summarizer(lecture,max_length=120, min_length=30)
  lectures_summaries[lecture] = summary
```

```python
import pandas as pd
df = pd.DataFrame(lectures_summaries.items(), columns=['Lecture', 'Summary'])
```
